# W4A9

```python
def loc_so_trung_nhau ( n ) :
    list_a = []
    while n > 0 :
        so_cuoi = n % 10
        list_a.append(so_cuoi)
        n = n // 10
    for i in range(len(list_a)) :
        for u in range(len(list_a))  :
            if i != u  :
                if list_a[i] == list_a [ u ] :
                    return False
    return True
n = int(input())
i = 0

list_b = []
if n ==  0  :
    print('no number')
else:
    while i* i <= n :
        so_cp = i * i
        if loc_so_trung_nhau(so_cp) == True :
            list_b.append(so_cp)
        i = i + 1
for i in range(len(list_b)) :
    print(list_b[i], end = " ")
```

# W4A10

```python
n = int(input())
max_len = 0
best_x = 1
for x in range(1, n + 1):
    current = x
    length = 1  
    while current != 1:
        if current % 2 == 0:
            current = current // 2
        else:
            current = current * 3 + 1  
        
        length += 1
    if length > max_len:
        max_len = length
        best_x = x
print(f"{best_x} {max_len}")
```

# W4A25

```python
max_val = None
min_val = None

print("Nhập dãy số nguyên (nhập -1 để dừng):")

while True:
    num = int(input("Nhập số: "))
    if num == -1:
        break
    if max_val is None:
        max_val = num
        min_val = num
    else:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num
if max_val is not None:
    print(f"Số lớn nhất là: {max_val}")
    print(f"Số nhỏ nhất là: {min_val}")
else:
    print("Bạn chưa nhập số nào ngoài số -1.")
```

# W4A31

```python
s = input("Nhập vào một chuỗi: ")
dem_hoa = 0
dem_thuong = 0
dem_so = 0
for i in range(len(s)):
    char = s[i]    
    if char.isupper():
        dem_hoa += 1
    elif char.islower():
        dem_thuong += 1
    elif char.isdigit():
        dem_so += 1
print(f"Số ký tự in hoa: {dem_hoa}")
print(f"Số ký tự in thường: {dem_thuong}")
print(f"Số ký tự chữ số: {dem_so}")
```
